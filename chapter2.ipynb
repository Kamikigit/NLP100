```python
#12 1列目をcol1.txtに，2列目をcol2.txtに保存
def print_data(filename, data):
    with open(filename, 'w') as f:
        f.write(data)
        
col1 = []
col2 = []
with open('/content/drive/My Drive/NLP100/data/popular-names.txt', 'r') as f:
    data = f.readlines()
    # dataを1行ずつ見てゆき、1列目をcol1、2列目をcol2とする
    col1 = [col.split()[0] + '\n' for col in data] 
    col2 = [col.split()[1] + '\n' for col in data]
    
print_data('/content/drive/My Drive/NLP100/data/col1.txt', ''.join(col1))
print_data('/content/drive/My Drive/NLP100/data/col2.txt', ''.join(col2))
```

```python
#13 col1.txtとcol2.txtをマージ
def read_data(filename):
    with open(filename, 'r') as f:
        data = f.read().split()
    return data

col1 = read_data('/content/drive/My Drive/NLP100/data/col1.txt')
col2 = read_data('/content/drive/My Drive/NLP100/data/col2.txt')
concat_data = []

for i in range(len(col1)):
    # ここでマージ
    concat_data.append(col1[i] + '\t')
    concat_data.append(col2[i] + '\n')
    
with open('/content/drive/My Drive/NLP100/data/result_13.txt', 'w') as f:
    f.write(''.join(concat_data))
```

```python
#19. 各行の1コラム目の文字列の出現頻度を求め，出現頻度の高い順に並べる
with open('/content/drive/My Drive/NLP100/data/popular-names.txt', 'r', encoding='utf-8') as f:
    data = f.readlines()
    # 重複削除をしてくれるsetを使う
    names = dict()

    #1行ずつ読み込んで、1列目をsetに追加する
    for dataline in data:
      name = dataline.split("\t")[0]
      if name in names.keys():  #辞書に既に存在したらvalueを+1
        names[name] += 1
      else:                     #辞書にない場合は登録する
        names[name] = 1

names = sorted(names.items(), key=lambda x:x[1]) # valueでソートする
print(names)
```
